# Law of Large Numbers - Coin Toss Example

### Description:
Consider a large number of coin tosses. Each of these coin tosses can be modelled as i.i.d Bernoulli random variables following some probability distribution. According to the Law of Large Numbers, the sample mean should converge to the expected mean as the number of samples gets larger. We plot the same here and verify that this is the case.

### Code: 

Imports

```python
import numpy as np
import random
import matplotlib.pyplot as plt 
```

Random Variable, its distribution and expectation

```python
X = [0,1]
f_X = [0.5,0.5] # Probability Mass Function
# Can change the values of X and f(X) and see how the LLN holds.
 
def expectation(random_variable,PMF):
    return np.sum(np.multiply(random_variable,PMF))
    
E_X = expectation(X,f_X)
print "E[X] =", E_X
```

Initialization of variables and lists

```python
N = 10000 # Maximum number of samples

num_samples_list = [] # 'i' samples 
sample_mean_list = [] 
abs_diff_list = [] 
 
sum_X_i = 0.0
sample_mean_i = 0
```

Calculations

```python
for i in range(1,N+1):
    
    num_samples_list.append(i)
    
    X_i = np.random.choice(X,p=f_X)
    sum_X_i = sum_X_i + X_i
    sample_mean_i = sum_X_i / i
    sample_mean_list.append(sample_mean_i) 
    
    abs_diff_list.append(abs(sample_mean_i-E_X))
```

Plots

```python
# Plot 1
plt.axhline(y=E_X, color='r')    
plt.plot(num_samples_list, sample_mean_list)
plt.title("Plot 1: Number of samples vs. Sample mean")
plt.xlabel("Number of samples (i)")
plt.ylabel("Sample mean")
plt.show() 
```

```python
# Plot 2
plt.plot(num_samples_list, abs_diff_list)
plt.title("Plot 2: Number of samples vs. Absolute difference error")
plt.xlabel("Number of samples (i)")
plt.ylabel("Absolute difference error")
plt.show() 
```

### Conclusion:

From Plot 1 we can see how the sample mean approaches the expected mean as number of samples gets larger. This is also evident in Plot 2 where the absolute difference between the two is plotted and it is consistently low as the number of samples gets larger. Thus, we can say LLN holds quite well which as just verified empirically.
